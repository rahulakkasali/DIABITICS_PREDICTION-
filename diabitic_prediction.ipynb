```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
X, y=load_diabetes(return_X_y=True)
X=X[:,[2]]
import pandas as pd
df=pd.DataFrame(X,columns=['BMI'])
df.head()
X_train ,X_test,y_train,y_test=train_test_split(X, y, test_size=20,shuffle=False)
df1=pd.DataFrame(X_train,columns=['BMI'])
df1.head()
df2=pd.DataFrame(X_test,columns=['BMI'])
df2.head()
df3=pd.DataFrame(y_train,columns=['target'])
df3.head()
df4=pd.DataFrame(y_test,columns=['target'])
df4.head()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)

df5=pd.DataFrame(y_pred,columns=['target'])
df5.head()
```

```python
# evaluate the model
from sklearn.metrics import mean_squared_error,r2_score
print(f"Mean squared error: {mean_squared_error(y_test, y_pred):.2f}")
print(f"Coefficient of determination: {r2_score(y_test, y_pred):.2f}")

```

```python
import matplotlib.pyplot as plt
plt.scatter(X_test,y_test,color='red',label='test')
plt.plot(X_test,y_pred,color='blue',label='predicted')
plt.title('BMI vs Diabetes Pedigree Function')
plt.xlabel('BMI')
plt.ylabel('Diabetes Pedigree Function')
plt.legend()

```

```python
plt.scatter(X_train,y_train,color='green',label='train')
plt.plot(X_train,model.predict(X_train),color='blue',label='predicted')
plt.title('BMI vs Diabetes Pedigree Function')
plt.xlabel('BMI')
plt.ylabel('Diabetes Pedigree Function')
plt.legend()
plt.show()

```
